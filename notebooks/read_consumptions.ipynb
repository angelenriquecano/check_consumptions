```python
import os
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
```

```python
dataDir = os.path.join(os.getcwd(), 'data')
print(os.listdir(dataDir))
```

```python
_DF = []

df = pd.read_csv(
    os.path.join(dataDir, os.listdir(dataDir)[2]),
    header=0,
    sep=';',
    decimal=','
)
df
    
```

```python
def fecha_hora_to_datetime(x):
    _str_hora = int(x['hora'].split(':')[0])
    return dt.datetime.strptime(f"{x['fecha']}-{_str_hora-1}:00", "%Y/%m/%d-%H:%M")
```

```python
df['Timestamp'] = df.apply(fecha_hora_to_datetime, axis=1)
df = df.set_index('Timestamp')
df['Pot_Contratada'] = 3.45
df
```

```python
import numpy as np
fig = go.Figure()
fig.add_traces(
    [
        go.Scatter(
            x = df.index,
            y = df.consumo_kWh,
            name='Consumo'               
        ),
        go.Scatter(
            x = df.index,
            y = df.Pot_Contratada,
            name='Energ_Max_Horaria = Pot_Contratada'               
        ),
        go.Bar(
            x = df.loc[df['metodoObtencion'] == 'Estimada'].index,
            y = df.loc[df['metodoObtencion'] == 'Estimada'].consumo_kWh.max() * np.ones(len(df.loc[df['metodoObtencion'] == 'Estimada'].index)),
            name='Estimadas',
            marker_color='rgb(255, 87, 51)', 
            marker_line_color='rgb(8,48,107)',
            marker_line_width=1.5, 
            opacity=0.3

        )
    ]
    
)
fig.update_layout(
    title='Consumo desde Enero 2024 hasta Octube 2024'
)
fig.show()
```
